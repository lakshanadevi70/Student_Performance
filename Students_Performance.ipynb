## Perform Multiple Linear Regression with Statsmodels

### Subtask:
Re-build and fit a multiple linear regression model using the prepared features (X) and 'FinalGrade' as the target variable (y) using `statsmodels` to get a detailed statistical summary.


**Reasoning**:
To perform multiple linear regression using `statsmodels` and obtain a detailed statistical summary, I will first import the `statsmodels.api` module. Then, I will add a constant to the feature DataFrame `X` for the intercept term, create an OLS model with the target variable `y` and the modified features, and finally fit the model.



```python
import statsmodels.api as sm

X_sm = sm.add_constant(X)
model_sm = sm.OLS(y, X_sm).fit()
```

**Reasoning**:
Now that the OLS model has been fitted using `statsmodels`, the next step is to display the detailed statistical summary of the model to analyze the regression results, as requested by the subtask.



```python
print(model_sm.summary())
```

### Multiple Linear Regression Analysis Summary

The provided `statsmodels` output presents the results of a multiple linear regression model built to predict `FinalGrade`.

**Overall Model Performance:**
*   **R-squared:** The model has an R-squared value of 0.008, and an Adjusted R-squared of -0.006. This indicates that only 0.8% of the variance in `FinalGrade` can be explained by the independent variables included in the model. An R-squared value close to zero suggests that the model is a poor fit for the data.
*   **F-statistic:** The F-statistic is 0.5707 with a corresponding p-value of 0.822. Since the p-value (0.822) is much greater than the conventional significance level of 0.05, we cannot reject the null hypothesis that all regression coefficients are equal to zero. This further supports the conclusion that the model, as a whole, is not statistically significant in predicting `FinalGrade`.

**Significance of Individual Predictors (p-values):**
We examine the p-values for each predictor to determine their statistical significance. A common threshold for significance is a p-value less than 0.05.

Based on the output:
*   **Gender:** p-value = 0.105 (not statistically significant at \alpha = 0.05)
*   **AttendanceRate:** p-value = 0.782 (not statistically significant)
*   **StudyHoursPerWeek:** p-value = 0.632 (not statistically significant)
*   **PreviousGrade:** p-value = 0.835 (not statistically significant)
*   **ExtracurricularActivities:** p-value = 0.981 (not statistically significant)
*   **ParentalSupport:** p-value = 0.691 (not statistically significant)
*   **Study Hours:** p-value = 0.422 (not statistically significant)
*   **Attendance (%):** p-value = 0.144 (not statistically significant)
*   **Online Classes Taken:** p-value = 0.686 (not statistically significant)

None of the individual independent variables are statistically significant predictors of `FinalGrade` at the 0.05 level.

**Estimated Impact (Coefficients):**
Although none of the predictors are statistically significant, here's an interpretation of their coefficients:
*   **Gender:** With a coefficient of 1.2154, if gender is coded as 1 for male and 0 for female, this suggests males might have a slightly higher `FinalGrade` by approximately 1.2 points, holding other variables constant. However, this is not statistically significant.
*   **AttendanceRate:** A coefficient of -0.0139 suggests a very slight negative relationship, meaning a 1-unit increase in `AttendanceRate` is associated with a 0.0139 decrease in `FinalGrade`, which is negligible and not significant.
*   **StudyHoursPerWeek:** A coefficient of 0.0279 indicates a very small positive relationship, implying that for every additional hour studied per week, the `FinalGrade` might increase by about 0.0279 points. This is not significant.

*   **ExtracurricularActivities:** A coefficient of -0.0087 suggests a negligible and non-significant negative association.

If 'Low' is 1, 'Medium' is 2, and 'High' is 3, this negative coefficient suggests that higher parental support (i.e., moving from 'Low' to 'Medium' or 'High') might be associated with a slightly lower `FinalGrade`, which is counterintuitive and not statistically significant.
*   **Study Hours:** A coefficient of 0.1962, suggests a modest positive relationship, meaning for every additional unit of 'Study Hours', `FinalGrade` might increase by about 0.1962 points. Not significant.

**Notes on Potential Issues:**
*   **Condition Number:** The summary reports a large condition number (2.17e+03). This is a strong indicator of potential multicollinearity, meaning some independent variables are highly correlated with each other. Multicollinearity can inflate the variance of the regression coefficients, making them unstable and difficult to interpret, and it can lead to high p-values even if the variables are truly related to the dependent variable. This could be a reason why no predictors appear significant.

**Conclusion:**
The multiple linear regression model, as currently specified, is not effective in predicting `FinalGrade` using the given features. Furthermore, none of the individual predictors show a statistically significant relationship with `FinalGrade`. The high condition number suggests that multicollinearity might be an issue, which could be obscuring true relationships between variables and should be investigated further if more robust analysis is desired.
